```python
##function to find the number of sylables
def syllable(word):
    letters=list(word)
    count =0
    v=['a','e','i','o','u']
    for i in letters:
                if i in v:
                    count=count+1

    
    if word.endswith('es') or word.endswith('ed'):
        count=count-1
    return count
```

## Accessing the urls and creating the text files

```python
for index, row in df.iterrows():
    # Extract the name and text
    file_name = row[0]
    url = row[1]
    
    # Create a valid file name by replacing any forbidden characters
    valid_file_name = ''.join(c for c in file_name if c.isalnum() or c in (' ', '_', '-')).rstrip()
    
    # Define the path to the text file
    text_file_path = f"{valid_file_name}.txt"
    
    # Write the text to the file
    page = requests.get(url).text
    doc=BeautifulSoup(page,"html.parser")
    page_text=doc.find(["div"],class_= "td-post-content tagdiv-type")
    if page_text:
     # Extract text from the element
     extract = page_text.get_text(strip=True)
     #print(extract)
    else:
     print("Element not found")
        
   
    lowercase=extract.lower()
    with open(text_file_path, 'w', encoding='utf-8') as text_file: #saving with file name equal to the url id
        text_file.write(lowercase)
    
    print(f"Saved {text_file_path}")

    
```

```python
# Removing punctuaation

# Function to remove punctuation from text
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

word_counts=[]
for file_name in file_list:
    # Read the file content in binary mode
    with open(file_name, 'rb') as file:
        file_content = file.read()
    
    # Decode the content to text
    text = file_content.decode('utf-8')  # Ensure you specify the correct encoding
    
    # Remove punctuation from the text
    text = remove_punctuation(text)

    # Remove leading and trailing whitespace
    paragraph = text.strip()
    # Split the paragraph into words based on whitespace
    words = paragraph.split()
    # Count the words
    word_count = len(words)
    
    
    #print(word_count)
    #print('\n')
    word_counts.append(word_count)

print(word_counts)
```

## Accessing the text file again to count the number of sentences and do other analysis

```python
#Declaring lists
Positive_Score=[]
Negative_Score=[]
Polarity_Score=[]
Subjectivity_Score=[]
Avg_Sentence_Length=[]
Percentage_of_Complex_Words=[]
Fog_Index=[]
Avg_Number_Of_Words_per_Sentence=[]
Complex_Word_count=[]
Word_Count=[]
Personal_Pronouns=[]
Average_Word_Length=[]
Syllable_Count_per_Word=[]

#accessing each text file
for file_name in file_list:
    sent_length=[]
    word_length=[]
    total_words=[]
    Pval=[]
    Nval=[]
    Pos=[]
    Neg=[]
    syl=[]
    Num_Syl=[]
    Avg_Syllable=0
    ch_tot=0
    
    Personal = ['i','we','my','ours','us']
    Person=[]
    
    #reading each text file
    with open(file_name, 'rb') as file:
        file_content = file.read()
        try:
          text = file_content.decode('utf-8')
        except UnicodeDecodeError:
          text = file_content.decode('latin-1')
        sent = sent_tokenize(text.strip()) #extracting each sentence to calculate the total number of sentences in each text
        sl=len(sent)
        sent_length.append(sl)
        wt=[]
        Pro=0
        
        for line in sent: #accessing each line to calculate word length
            words=line.split()
            wl=len(words)
            wt.append(wl)
        word_length.append(wt)
       
        total = np.sum([num for num in word_length])
        
        for line in sent:#to calculate number of characters in each word
            words=line.split()
            for word in words: 
              ch=tot_char(word)
              ch_tot=ch_tot+ch

        for line in sent:# to find the positive and negative words
            words=line.split()
            for word in words:
              if word in Dict["positive"]:
                P =1
                Pval.append(P)
              elif word in Dict["negative"]:
                N =-1
                Nval.append(N)
              Pt=np.sum([p for p in Pval])
              Nt=np.sum([n for n in Nval])*-1
        Positive_Score.append(Pt)
        Negative_Score.append(Nt)

        
        
        for line in sent:
            words=line.split()
            for word in words: #to find the number of personal pronouns
              if word in Personal:
                 Pro=Pro+1
                
                
            count=0       
            for word in words:# to find the number of syllables
              letters=list(word)
              leng=len(letters)
              count=syllable(word)
              Nu=count/leng
              Num_Syl.append(Nu)
              
            if count>2:
               syl.append(word)
                
    ls=len(syl)
    
    Avg_Syllable=np.mean([p for p in Num_Syl])  #calculating average number of syllables
    
    Syllable_Count_per_Word.append(Avg_Syllable)
    
    Polarity_score=(Pt-Nt)/((Pt+Nt)+0.000001)#calculating polarity scores
   
    Subjective_score=(Pt+Nt)/(total+0.000001)#calculating subjective scores
    
    Polarity_Score.append(Polarity_score)
    
    Subjectivity_Score.append(Subjective_score)
    
    Average_Sentence_length= total/sent_length#calculating average sentence length
    Avg_Sentence_Length.append( Average_Sentence_length)

    AverageNoWordsperSentence=total/sent_length#calculating average number of words per sentence
    Avg_Number_Of_Words_per_Sentence.append(AverageNoWordsperSentence)
    
    PercentageofComplexwords=ls/total #Calculating Readability
    Percentage_of_Complex_Words.append(PercentageofComplexwords)
    Average_Word_Length.append(Average_word_Length)
    Complex_Word_count.append(ls)
    
    Fog_index= 0.4*(Average_Sentence_length+PercentageofComplexwords)#calculating fog index
    Fog_Index.append(Fog_index)
    Average_word_Length=ch_tot/total
    Word_Count.append(total)

    #Now the values for each text file is printed for reference
    # Pos.append(Pval)
    print(f"Positive Valu={Pt}")
   
    # Neg.append(Nval)
    #Nt=np.abs(Nt)
    print(f"Negative Value={Nt}")
    Personal_Pronouns.append(Pro)
    print(f"Number of Personal Pronouns={Pro}")
    print(f"Polarity Score={Polarity_score}")
    
    
    print(f"Subjective Score={Subjective_score}")
    
   #Analysis of Readability
    print(f"Average Sentence length={Average_Sentence_length}")
    print(f"Number of Complex words={ls}")
    
    print(f"Percentage of Complex words={PercentageofComplexwords}")
    print(f"Fog Index={Fog_index}")
    print(f"Average No Words per Sentence={AverageNoWordsperSentence}")
    
    print(f"Sentence Length={sent_length}")
    print(f"Word Length={word_length}")
    print(f"Total Number of words={total}")
    print(f"Total Number of Characters={ch_tot}")
    print(f"Average Word Length={Average_word_Length}")
    #total_words.append(total)
    print(f"SyllableCountPerWord={Syllable_Count_per_Word}")
    print("\n")
    
```
